1. Импорт библиотеки NumPy с общепринятым сокращением

```python
# стандартный импорт numpy
import numpy as np
```

2. Примеры создания массивов

```python
# создания массивов из обычных списков
arr_list1 = np.array([1, 2, 3])
arr_list2 = np.array(list((4, 5, 6)))
# создания массивов с использованием методов numpy
arr_zeros = np.zeros((3, 5)) # матрица нулей
arr_ones = np.ones(3) # матрица единиц
arr_eye = np.eye(4) # единичная матрица 
arr_arange = np.arange(1, 11, 2) # массив от 1 до 11 с шагом 2
arr_empty = np.empty((3, 2)) # массив с "мусорными" значения
arr_list2, arr_eye
```

3. Демонстрация атрибутов

```python
# вывод формы массивов
arr_zeros.shape, arr_ones.shape
```

```python
# вывод типа данных массивов
arr_arange.dtype, arr_empty.dtype
```

4. Примеры индексирования и срезов

```python
# получение отдельного элемента;
arr_list1[0], arr_eye[0,0]
```

```python
# извлечение подмассива с помощью среза
arr_arange[2:5], arr_eye[2:4,1:4]
```

5. Примеры векторных операций и базовых статистик

```python
# поэлементное сложение или умножение массивов
arr_arange + np.arange(3, 12, 2), 2 * arr_ones * np.eye(3)
```

```python
# вычисление простой статистики
arr_random = np.random.randint(0,5,(2,3))
print(f"""Случайный массив:\n{arr_random}
Среднее: {np.mean(arr_random)}
Медиана: {np.median(arr_random)}
Стандартное отклонение: {np.std(arr_random)}
Дисперсия: {np.var(arr_random)}""")
```
